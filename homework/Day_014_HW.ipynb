```python
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.random.randint(0 ,50, 1000)
y = np.random.randint(0, 50, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

```python
x = np.random.randint(0 ,50, 1000)
y= x+np.random.normal(0, 10, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

```python
x = np.random.randint(0 ,50, 1000)
y= -x+np.random.normal(0, 10, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```
