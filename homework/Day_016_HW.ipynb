```python
import os 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
dir_data = 'data/'
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

```python
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

```python
age_data = app_train[['TARGET', 'DAYS_BIRTH']]
age_data['YEARS_BIRTH'] = age_data['DAYS_BIRTH']/365
```

```python
age_data.head()
```

```python
np.linspace(20, 70, 11)
```

```python
bin_cut = np.linspace(20, 70, 11)
```

```python
age_data['YEARS_BINNED'] = pd.cut(age_data['YEARS_BIRTH'], bins = bin_cut)
```

```python
age_data['YEARS_BINNED'].head()
```

```python
print(age_data['YEARS_BINNED'].value_counts())
age_data.head()
```

```python
year_group_sorted = age_data['YEARS_BINNED'].sort_values(ascending = True).values
```

```python
plt.figure(figsize = (8, 6))
for i in range(len(year_group_sorted)):
    sns.distplot(age_data.loc[(age_data['YEARS_BINNED'] == year_group_sorted[i]) & \
                              (age_data['TARGET'] == 0), 'YEARS_BIRTH'], label = str(year_group_sorted[i]))
    
    sns.distplot(age_data.loc[(age_data['YEARS_BINNED'] == year_group_sorted[i]) & \
                              (age_data['TARGET'] == 1), 'YEARS_BIRTH'], label = str(year_group_sorted[i]))
plt.title('KDE with Age groups')
plt.show()
```

```python
age_groups = age_data.groupby('YEARS_BINNED').mean()
age_groups
```

```python
plt.figure(figsize = (8, 8))

px = age_data['YEARS_BINNED'].values
py = age_data['TARGET'].values

sns.barplot(px, py)

plt.xticks(rotation = 70)
plt.xlabel('Age Group (years)')
plt.ylabel('Failure to Repay (%)')
plt.title('Failure to Repay by Age Group')
```
